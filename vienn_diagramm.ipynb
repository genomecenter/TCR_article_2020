```python
#функция чтения первых N клонотипов из файла, заданного путем, в сет
#пусть длину файла выдает тоже

def read_clon_file(path):
    
    with open(path) as f:
        lis = [line.split('\t') for line in f] 
        lenlis=len(lis)
        arr=[x[3] for x in lis]
        
        ff=int(lenlis/20)
        arr=arr[0:ff]
        
    return int(lenlis/20), arr
    #return lenlis, arr
```

```python
#чтение клонотипов альфа 5 групп из соответствующих 5 файлов

pathCl01Agr="./PoolSamples/PoolSamples05092019alpha.pool.aa.table.txt"
pathCl02Agr="./PoolSamples/PoolSamples18112019alpha.pool.aa.table.txt" 
pathCl03Agr="./PoolSamples/PoolSamples16012020alpha.pool.aa.table.txt"
pathCl04Agr="./PoolSamples/PoolSamples27112019_1-5_alpha.pool.aa.table.txt"
pathCl05Agr="./PoolSamples/PoolSamples27112019_6-10_alpha.pool.aa.table.txt"


gr1A=read_clon_file(pathCl01Agr)
print('gr_1 ',gr1A[0])
gr2A=read_clon_file(pathCl02Agr)
print('gr_2 ',gr2A[0])
gr3A=read_clon_file(pathCl03Agr)
print('gr_3 ',gr3A[0])
gr4A=read_clon_file(pathCl04Agr)
print('gr_4 ',gr4A[0])
gr5A=read_clon_file(pathCl05Agr)
print('gr_5 ',gr5A[0])


```

```python
#построить диаграмму альфа

ar1=set(gr1A[1])
ar2=set(gr2A[1])
ar3=set(gr3A[1])
ar4=set(gr4A[1])
ar5=set(gr5A[1])


clonotypes = {
    "Group 1 alpha": ar1,
    "Group 2 alpha": ar2,
    "Group 3 alpha": ar3,
    "Group 4 alpha": ar4,
    "Group 5 alpha": ar5,
}
from venn import venn
venn(clonotypes)
```

```python
#чтение клонотипов бета 5 групп из соответствующих 5 файлов

pathCl01Bgr="./PoolSamples/PoolSamples05092019beta.pool.aa.table.txt"   
pathCl02Bgr="./PoolSamples/PoolSamples18112019beta.pool.aa.table.txt"
pathCl03Bgr="./PoolSamples/PoolSamples16012020beta.pool.aa.table.txt" 
pathCl04Bgr="./PoolSamples/PoolSamples27112019_1-5_beta.pool.aa.table.txt"
pathCl05Bgr="./PoolSamples/PoolSamples27112019_6-10_beta.pool.aa.table.txt"

gr1B=read_clon_file(pathCl01Bgr)
print('gr_1 ',gr1B[0])
gr2B=read_clon_file(pathCl02Bgr)
print('gr_2 ',gr2B[0])
gr3B=read_clon_file(pathCl03Bgr)
print('gr_3 ',gr3B[0])
gr4B=read_clon_file(pathCl04Bgr)
print('gr_4 ',gr4B[0])
gr5B=read_clon_file(pathCl05Bgr)
print('gr_5 ',gr5B[0])
```

```python
#построить диаграмму бета
ar1=set(gr1B[1])
ar2=set(gr2B[1])
ar3=set(gr3B[1])
ar4=set(gr4B[1])
ar5=set(gr5B[1])


clonotypes = {
    "Group 1 beta": ar1,
    "Group 2 beta": ar2,
    "Group 3 beta": ar3,
    "Group 4 beta": ar4,
    "Group 5 beta": ar5,
}

#import matplotlib
from venn import venn
venn(clonotypes)
```

```python
#сравнение 4 и 5 групп


gr4A=read_clon_file(pathCl04Agr)
print('gr_4 ',gr4A[0])
gr5A=read_clon_file(pathCl05Agr)
print('gr_5 ',gr5A[0])

ar4=set(gr4A[1])
ar5=set(gr5A[1])

clonotypes = {
    "Group 4 alpha": ar4,
    "Group 5 alpha": ar5,
}
from venn import venn
venn(clonotypes)
```

```python
#https://stackoverflow.com/questions/46144452/writing-a-set-to-an-output-file-in-python
set4minus5=ar4.difference(ar5)
len(set4minus5)

with open('top5pr_4_minus_top5pr_5_alpha.txt','w') as f:
    f.write(str(set4minus5))
f.close()
```
