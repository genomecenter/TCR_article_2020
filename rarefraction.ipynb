```python
def read_rar_file(path):
    
    with open(path) as f:
        lis = [line.split('\t') for line in f] 
        lenlis=len(lis)
        xval=[float(x[2]) for x in lis]
        yval=[float(x[3]) for x in lis]
        ymin=[float(x[4]) for x in lis]
        ymax=[float(x[5]) for x in lis]
        
       
    return xval,yval,ymin,ymax
```

```python
first=read_rar_file(pathL03)
second=read_rar_file(pathL04)
third=read_rar_file(pathL05)
```
